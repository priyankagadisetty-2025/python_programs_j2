# Iterations in python 
1. iterations are used to work on repeatative tasks or repeatative performances
2. python having 2 types of iterations
3. `for` loop
4. `while` loop 

```python - for loop 
for iter_variable in sequence:
    print("for is keyword to start the iterations")
    print("iter_variables is to get values from sequence")
    print("in is keywords to valid sequence presence")
    print("sequence is collection like range,string, list,tuple...etc")
```

```python - while loop
while (condition):
    print("while is keywords to valid condtion based on data orvalues")
    print("if conditions is true then return inside statements")
```

````python
# write a program to print 1 to 20 numbers using for loop
for i in range(1,21):
    print(i)
````

````python
# write a program to iterae string for loop 
info="welcome to python program"
for char in info:
    print(char)
````

````python
# write a program to print even numbers in between 50 to 100 using for loop?
for n in range(50,101):
    if (n%2 == 0):
        print(n,end=",")
````

````python
# write a program to print odd numbers in between 50 to 100 using for loops?
for n in range(50,100):
    if (n%2 ==1):
        print(n,end=",")
````

````python
# write a program to find the vowels from string using for loop?
data="python programming"
for char in data:
    if (char in "aeiouAEIOU"):
       print(char)
````

````python
# write a program to find the non-vowels from string using for loop?
data="python programming"
for char in data:
    if (char not in "aeiouAEIOU"):
        print(char)
````
